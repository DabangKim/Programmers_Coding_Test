```python
def solution(cacheSize, cities):
    cache = []
    cnt = 0
    for city in cities:
        city = city.lower()
        if city not in cache:
            cache.append(city)
            cnt += 5
            if len(cache) > cacheSize:
                cache.pop(0)
        else:
            cache.append(city)
            cnt += 1
            cache.remove(city)
            if len(cache) > cacheSize:
                cache.pop(0)
    return cnt
```
