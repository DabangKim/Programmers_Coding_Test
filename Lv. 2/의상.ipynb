```python
def solution(clothes):
    # 옷 종류 확인
    collection = [j for i,j in clothes]
    collection = list(set(collection))

    # 옷 종류별 의상 이름
    name = [[i for i,j in clothes if j==collect] for collect in collection]

    # 정답
    answer = [len(na) for na in name]
    n = len(answer)
    total = 0

    for i in range(1, 1 << n):
        product = 1
        for j in range(n):
            if i & (1 << j):
                product *= answer[j]
        total += product

    return total
```

```python
from collections import Counter

def solution(clothes):
    counter = Counter([kind for _, kind in clothes])

    answer = 1
    for count in counter.values():
        answer *= (count + 1)

    return answer - 1
```
