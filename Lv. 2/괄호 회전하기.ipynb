```python
def is_valid(s):
    stack = []
    bracket_map = {')': '(', ']': '[', '}': '{'}

    for char in s:
        if char in '([{':
            stack.append(char)
        elif char in ')]}':
            if not stack or stack[-1] != bracket_map[char]:
                return False
            stack.pop()
    return not stack

def solution(s):
    count = 0
    n = len(s)

    for i in range(n):
        rotated = s[i:] + s[:i]
        if is_valid(rotated):
            count += 1

    return count
```
