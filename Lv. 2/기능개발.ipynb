```python
def solution(progresses, speeds):

    # 몇 일 후에 배포 가능한지 각각 계산
    check = []
    for i in range(len(progresses)):
        che = (100 - progresses[i]) % speeds[i]
        if che == 0:
            check.append((100 - progresses[i]) // speeds[i])
        else:
            check.append((100 - progresses[i]) // speeds[i] + 1)

    # 정답
    checker = check[0]
    answer = []

    for j in range(len(check)):
        if check[j] > checker:
            answer.append(len(check[check.index(checker):j]))
            checker = check[j]
            if j+1 == len(check):
                answer.append(1)
        elif check[j] <= checker and j+1 == len(check):
            answer.append(len(check[check.index(checker):]))

    return answer
```
