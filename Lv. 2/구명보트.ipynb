```python
def solution(people, limit):
    people.sort()
    left = 0
    right = len(people) - 1
    boats = 0

    while left <= right:
        if people[left] + people[right] <= limit:
            left += 1  # 둘이 같이 탈 수 있으면 가벼운 사람도 이동
        right -= 1  # 무거운 사람은 항상 이동
        boats += 1  # 보트 1대 사용

    return boats
```
