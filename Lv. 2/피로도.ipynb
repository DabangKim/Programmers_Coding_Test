```python
from itertools import permutations

def solution(k, dungeons):
    cnt = []
    all_dungeons = list(permutations(dungeons))
    for all_dungeon in all_dungeons:
        count = 0
        false_k = k
        for i,j in all_dungeon:
            if false_k >= i:
                false_k -= j
                count += 1
        cnt.append(count)
    return max(cnt)
```
