```python
def solution(s):
    # 맨 앞과 맨 뒤 중괄호 제거
    s = s[2:-2]

    # 중괄호 단위로 분리
    parts = s.split("},{")
    for i in range(len(parts)):
        parts[i] = parts[i].split(',')

    # 길이 순서대로 정렬
    sorted_parts = sorted(parts, key=len)

    # 정답
    answer = []
    for part in sorted_parts:
        for num in part:
            if int(num) not in answer:
                answer.append(int(num))
    return answer
```
