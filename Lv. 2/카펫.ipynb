```python
def solution(brown, yellow):
    temp = []
    brown -= 4
    for i in range(1, yellow//2+2):
        if yellow % i == 0:
            temp_2 = [max(i, yellow//i), min(i, yellow//i)]
            if temp_2 not in temp:
                temp.append(temp_2)
    for ans in temp:
        if sum(ans)*2 == brown:
            ans[0] += 2
            ans[1] += 2
            return ans
```
