```python
# 시간 초과

def solution(s):
    s = list(s)
    for i in range(len(s)):
        if s[i] == '(':
            for j in range(i+1, len(s)):
                if s[j] == ')':
                    s[i] = ''
                    s[j] = ''
                    break
    print(s)
    return True if ''.join(s) == '' else False
```

```python
def solution(s):
    stack = []
    for ch in s:
        if ch == '(':
            stack.append(ch)
        else:  # ch == ')'
            if not stack:
                return False
            stack.pop()
    return not stack  # 스택이 비어 있으면 True, 아니면 False
```
