```python
def solution(elements):
    answer = []
    for i in range(1, len(elements)+1):
        temp = []
        for j in range(len(elements)-i+1):
            temp.append(sum(elements[j:j+i]))
        elements.append(elements[i-1])
        temp = list(set(temp))
        for i in temp:
            answer.append(i)
    return len(list(set(answer)))
```

```python
def solution(elements):
    n = len(elements)
    extended = elements * 2
    sums = set()

    for size in range(1, n + 1):
        current_sum = sum(extended[:size])
        sums.add(current_sum)

        for i in range(1, n):
            current_sum = current_sum - extended[i - 1] + extended[i + size - 1]
            sums.add(current_sum)

    return len(sums)
```
