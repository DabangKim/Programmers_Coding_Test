```python
# 시간 초과
def solution(topping):
    li = []
    answer = 0

    while topping != []:
        top = topping.pop(0)
        li.append(top)
        if len(set(li)) == len(set(topping)):
            answer += 1

    return answer
```

```python
# 시간 초과
def solution(topping):
    answer = 0
    for i in range(1, len(topping)):
        if len(set(topping[:i])) == len(set(topping[i:])):
            answer += 1
    return answer
```

```python
def solution(topping):
    answer = 0
    left = {}
    right = {}

    # 먼저 전체 토핑 빈도수를 right에 저장
    for t in topping:
        if t in right:
            right[t] += 1
        else:
            right[t] = 1

    for i in range(len(topping) - 1):  # 끝까지 자를 수는 없으니 -1까지
        t = topping[i]

        # 왼쪽에 토핑 추가
        if t in left:
            left[t] += 1
        else:
            left[t] = 1

        # 오른쪽에서 토핑 제거
        right[t] -= 1
        if right[t] == 0:
            del right[t]

        # 공평하게 나뉘었는지 확인
        if len(left) == len(right):
            answer += 1

    return answer
```
