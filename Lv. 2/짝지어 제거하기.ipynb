```python
# 시간 초과

def solution(s):
    s = list(s)
    while len(s) != 0:
        for_change = len(s)
        for i in range(1, len(s)):
            if s[i-1] == s[i]:
                del s[i-1]
                del s[i-1]
                break
        if len(s) == for_change:
            return 0
    return 1
```

```python
def solution(s):
    stack = []
    for char in s:
        if stack and stack[-1] == char:
            stack.pop()
        else:
            stack.append(char)
    return 1 if not stack else 0
```
