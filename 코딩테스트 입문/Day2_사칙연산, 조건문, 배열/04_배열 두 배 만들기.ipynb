```python
def solution(numbers):
    for i in range(len(numbers)):
        numbers[i] *= 2
    return numbers
```

```python
# Pythonic한 방식

def solution(numbers):
    return [num * 2 for num in numbers]
```

```python
# 프로그래머스러운 방식

def solution(numbers):
    return list(map(lambda x: x * 2, numbers))
```
