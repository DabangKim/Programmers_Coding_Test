```python
def divisor(num):  # 공약수
    temp = []
    for i in range(1, num+1):
        if num % i == 0:
            temp.append(i)
    return temp

def GCD(num1, num2):  # 최대공약수
    temp1 = divisor(num1)
    temp2 = divisor(num2)
    temp3 = []
    for i in temp1:
        if i in temp2:
            temp3.append(i)
    return temp3[-1]

def LCM(num1, num2):  # 최소공배수
    temp = GCD(num1, num2)
    return int(num1 / temp) * int(num2 / temp) * temp

def solution(numer1, denom1, numer2, denom2):
    temp1 = LCM(denom1, denom2)
    temp2 = numer1 * int(temp1 / denom1)
    temp3 = numer2 * int(temp1 / denom2)
    temp4 = temp2 + temp3
    if GCD(temp1, temp4):
        return [int(temp4 / GCD(temp1, temp4)), int(temp1 / GCD(temp1, temp4))]
    else:
        return [temp4, temp1]
```
