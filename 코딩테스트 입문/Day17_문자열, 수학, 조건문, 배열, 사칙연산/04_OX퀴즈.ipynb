```python
def solution(quiz):
    answer = []
    for i in quiz:
        temp = i.split()
        temp_answer = eval(temp[0]+temp[1]+temp[2])
        if temp_answer == int(temp[-1]):
            answer.append("O")
        else:
            answer.append("X")
    return answer
```

```python
def valid(equation):
    equation = equation.replace('=', '==')
    return eval(equation)

def solution(equations):
    return ["O" if valid(equation) else "X" for equation in equations]
```
