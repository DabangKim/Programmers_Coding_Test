```python
def solution(array):
    temp_1 = []  # 중복되는 원소는 하나로
    temp_2 = []  # 원소 개수 세기
    temp_3 = []  # 원소 개수만
    for i in array:
        if i not in temp_1:
            temp_1.append(i)
    for j in temp_1:
        temp_2.append((j, array.count(j)))
        temp_3.append(array.count(j))
    if temp_3.count(max(temp_3)) != 1:
        return -1
    else:
        for num in range(len(temp_2)):
            if temp_2[num][1] == max(temp_3):
                return temp_2[num][0]
```
