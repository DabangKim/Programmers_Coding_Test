```python
def divisor(num):  # 공약수
    return [i for i in range(1, num+1) if num % i == 0]

def GCD(num1, num2):  # 최대공약수
    temp1 = divisor(num1)
    temp2 = divisor(num2)
    return [i for i in temp1 if i in temp2][-1]

def solution(a, b):
    denominator = b // GCD(a, b)  # 기약분수의 분모
    while denominator % 2 == 0:
        denominator //= 2
    while denominator % 5 == 0:
        denominator //= 5
    return 1 if denominator == 1 else 2
```
