```python
def solution(lines):
    # 길이순으로 정렬
    lines_sorted = sorted(lines, key=lambda line: line[1] - line[0])

    # 선분 2개씩 비교
    temp1 = [lines_sorted[0], lines_sorted[1]]
    temp2 = [lines_sorted[1], lines_sorted[2]]
    temp3 = [lines_sorted[0], lines_sorted[2]]

    # 케이스별 분류 -> 겹치는 부분 common에 추가
    temp = [temp1, temp2, temp3]
    common = []
    for i in temp:
        if i[1][1] > i[0][0] and i[1][1] < i[0][1]:
            common.append([ i[0][0] , i[1][1] ])
        elif i[0][0] >= i[1][0] and i[0][1] <= i[1][1]:
            common.append([ i[0][0] , i[0][1] ])
        elif i[1][0] > i[0][0] and i[1][0] < i[0][1]:
            common.append([ i[1][0] , i[0][1] ])

    # 겹치는 부분이 몇 개인지에 따라서 분류
    if len(common) == 0:
        return 0
    elif len(common) == 1:
        return common[0][1] - common[0][0]
    elif len(common) == 2:
        return common[0][1] - common[0][0] + common[1][1] - common[1][0]
    else:
        min_max = []
        for i in common:
            min_max.append(i[0])
            min_max.append(i[1])
        return max(min_max) - min(min_max)
```
