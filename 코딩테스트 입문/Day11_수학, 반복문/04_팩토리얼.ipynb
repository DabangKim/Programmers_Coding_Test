```python
def solution(n):
    num = 1
    while True:
        answer = 1
        for i in range(1, num+1):
            answer *= i
        if answer > n:
            return num-1
            break
        else:
            num += 1
```

```python
def solution(n):
    answer = 1
    factorial = 1
    while factorial <= n:
        answer += 1
        factorial = factorial * answer
    answer -= 1
    return answer
```
