```python
def solution(common):
    temp = [common[i+1] - common[i] for i in range(len(common) - 1)]
    if temp.count(temp[0]) == len(temp):
        return common[-1] + temp[0]
    else:
        return common[-1] * int(common[1] / common[0])
```
