```python
# num에 따라 연속된 정수 만드는 함수
def maker(num):
    answer = []
    for i in range(num):
        temp = []
        for j in range(num):
            temp.append(j-i)
        answer.append(temp)
    return answer

# 정답
def solution(num, total):
    temp = maker(num)
    plus = 0
    for i in temp:
        if (total - sum(i)) % num == 0:
            plus = (total - sum(i)) // num
            for j in range(len(i)):
                i[j] += plus
            return i
```
