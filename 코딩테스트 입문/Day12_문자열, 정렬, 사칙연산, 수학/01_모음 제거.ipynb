```python
def solution(my_string):
    answer = ''
    for a in my_string:
        if a == 'a' or a == 'e' or a == 'i' or a == 'o' or a == 'u':
            pass
        else:
            answer += a
    return answer
```

```python
def solution(my_string):
    vowels = ['a','e','i','o','u']
    for vowel in vowels:
        my_string = my_string.replace(vowel, '')
    return my_string
```
