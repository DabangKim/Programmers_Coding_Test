```python
def solution(spell, dic):
    for i in dic:
        temp = []
        for j in spell:
            if j in i:
                temp.append(True)
            else:
                temp.append(False)
        if False not in temp:
            return 1
    if False in temp:
        return 2
```

```python
def solution(spell, dic):
    spell = set(spell)
    for s in dic:
        if not spell-set(s):
            return 1
    return 2
```
