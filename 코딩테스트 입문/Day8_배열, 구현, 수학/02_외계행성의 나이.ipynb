```python
def solution(age):
    temp = list(enumerate(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']))
    answer = ''
    for i in str(age):
        for j in range(len(temp)):
            if i == str(temp[j][0]):
                answer += temp[j][1]
    return answer
```
