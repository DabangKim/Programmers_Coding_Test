```python
def solution(dots):
    if dots[0][0] == dots[1][0]:
        a = abs(dots[0][1] - dots[1][1])
    elif dots[0][0] == dots[2][0]:
        a = abs(dots[0][1] - dots[2][1])
    else:
        a = abs(dots[0][1] - dots[3][1])

    if dots[0][1] == dots[1][1]:
        b = abs(dots[0][0] - dots[1][0])
    elif dots[0][1] == dots[2][1]:
        b = abs(dots[0][0] - dots[2][0])
    else:
        b = abs(dots[0][0] - dots[3][0])

    return a*b
```
