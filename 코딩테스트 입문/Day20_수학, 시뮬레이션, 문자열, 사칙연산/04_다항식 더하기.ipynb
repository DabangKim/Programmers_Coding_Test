```python
def solution(polynomial):
    temp_x = []
    temp_xn = 0
    temp_n = 0
    for i in polynomial.split():
        if 'x' in i:
            temp_x.append(i)
        elif i.isnumeric():
            temp_n += int(i)

    for j in temp_x:
        if j == 'x':
            j = '1x'
        temp_xn += int(list(j)[0])

    if temp_xn != 0 and temp_n != 0:
        return str(temp_xn) + 'x + ' + str(temp_n)
    elif temp_xn == 0 and temp_n != 0:
        return str(temp_n)
    elif temp_xn != 0 and temp_n == 0:
        return str(temp_xn) + 'x'
```

```python
def solution(polynomial):
    constant = 0
    coefficient = 0
    temp = []
    for i in polynomial.split():
        if i.isdigit():
            constant += int(i)
        elif 'x' in i:
            temp.append(i)

    for j in temp:
        if j =='x':
            j = '1x'
        coefficient += int(j.replace('x', ''))

    if constant == 0:
        if coefficient == 1:
            return 'x'
        return str(coefficient) + 'x'
    elif coefficient == 0:
        return str(constant)
    else:
        if coefficient == 1:
            return 'x + ' + str(constant)
        return str(coefficient) + 'x + ' + str(constant)
```

```python
def solution(polynomial):
    xnum = 0
    const = 0
    for c in polynomial.split(' + '):
        if c.isdigit():
            const+=int(c)
        else:
            xnum = xnum+1 if c=='x' else xnum+int(c[:-1])
    if xnum == 0:
        return str(const)
    elif xnum==1:
        return 'x + '+str(const) if const!=0 else 'x'
    else:
        return f'{xnum}x + {const}' if const!=0 else f'{xnum}x'
```
