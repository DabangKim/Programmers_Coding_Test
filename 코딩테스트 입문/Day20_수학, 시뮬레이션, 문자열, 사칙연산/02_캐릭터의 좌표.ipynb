```python
def solution(keyinput, board):
    ans = [0,0]
    for key in keyinput:
        if key == 'up':
            ans[1] += 1
            if ans[1] > board[1]//2:
                ans[1] = board[1]//2
        elif key == 'down':
            ans[1] -= 1
            if ans[1] < -(board[1]//2):
                ans[1] = -(board[1]//2)
        elif key == 'right':
            ans[0] += 1
            if ans[0] > board[0]//2:
                ans[0] = board[0]//2
        else:
            ans[0] -= 1
            if ans[0] < -(board[0]//2):
                ans[0] = -(board[0]//2)
    return ans
```
