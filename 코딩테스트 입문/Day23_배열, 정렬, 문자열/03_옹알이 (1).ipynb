```python
def solution(babbling):
    pronounce = ["aya", "ye", "woo", "ma"]
    for i in range(len(babbling)):
        for word in pronounce:
            if word in babbling[i]:
                babbling[i] = babbling[i].replace(word, " ")
    for j in range(len(babbling)):
        while len(babbling[j]) > 0:
            if ' ' in babbling[j]:
                babbling[j] = babbling[j].replace(' ', '')
            else:
                break
    return babbling.count('')
```
