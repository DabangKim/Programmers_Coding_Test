```python
def solution(numlist, n):
    temp = [i-n for i in numlist]
    n_temp = list(zip(numlist, temp))
    n_temp.sort(key = lambda x : abs(x[1]))
    for i in range(len(n_temp)-1):
        if abs(n_temp[i][1]) == abs(n_temp[i+1][1]) and n_temp[i][1] < n_temp[i+1][1]:
            n_temp[i], n_temp[i+1] = n_temp[i+1], n_temp[i]
    return [i[0] for i in n_temp]
```
