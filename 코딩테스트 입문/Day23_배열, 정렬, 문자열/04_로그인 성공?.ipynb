```python
def solution(id_pw, db):
    temp = []
    for info in db:
        if info[0] == id_pw[0] and info[1] == id_pw[1]:
            temp.append("login")
        elif info[0] == id_pw[0] and info[1] != id_pw[1]:
            temp.append("wrong pw")
        else:
            temp.append("fail")
    if 'login' in temp:
        return 'login'
    if 'wrong pw' in temp:
        return 'wrong pw'
    if 'fail' in temp:
        return 'fail'
```
