```python
def solution(score):
    avg_scores = [(i, (s + t) / 2) for i, (s, t) in enumerate(score)]
    sorted_scores = sorted(avg_scores, key=lambda x: x[1], reverse=True)

    rank = [0] * len(score)
    current_rank = 1

    for i in range(len(sorted_scores)):
        if i > 0 and sorted_scores[i][1] != sorted_scores[i - 1][1]:
            current_rank = i + 1
        rank[sorted_scores[i][0]] = current_rank

    return rank
```
