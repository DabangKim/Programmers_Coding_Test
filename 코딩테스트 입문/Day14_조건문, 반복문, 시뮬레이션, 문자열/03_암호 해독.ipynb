```python
def solution(cipher, code):
    answer = ''
    for a in range(code-1, len(cipher), code):
        answer += cipher[a]
    return answer
```

```python
def solution(cipher, code):
    return ''.join(cipher[i] for i in range(code-1, len(cipher), code))
```
