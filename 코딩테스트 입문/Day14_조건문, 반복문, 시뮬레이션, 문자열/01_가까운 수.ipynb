```python
def solution(array, n):
    temp = []
    temp_abs = []
    for i in array:
        temp.append(i-n)
        temp_abs.append(abs(i-n))
    if temp_abs.count(min(temp_abs)) != 1:
        for num in temp:
            if num == -min(temp_abs):
                return array[temp.index(num)]
        for num in temp:
            if num == min(temp_abs):
                return array[temp.index(num)]
    else:
        return array[temp_abs.index(min(temp_abs))]
```
