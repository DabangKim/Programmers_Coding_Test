```python
def solution(cards1, cards2, goal):
    check = []
    for word in goal:
        if cards1 != [] and word == cards1[0]:
            check.append(True)
            cards1.pop(0)
        elif cards2 != [] and word == cards2[0]:
            check.append(True)
            cards2.pop(0)
        else:
            check.append(False)
    if False in check:
        return "No"
    else:
        return "Yes"
```
