```python
def solution(n, arr1, arr2):
    # 2진수로 변환
    for i in range(n):
        arr1[i] = format(arr1[i], 'b')
        if len(arr1[i]) < n:
            arr1[i] = '0' * (n - len(arr1[i])) + arr1[i]

        arr2[i] = format(arr2[i], 'b')
        if len(arr2[i]) < n:
            arr2[i] = '0' * (n - len(arr2[i])) + arr2[i]

    # 정답
    answer = []
    for i in range(n):
        temp = ''
        for j in range(n):
            if arr1[i][j] == '1' or arr2[i][j] == '1':
                temp += '#'
            elif arr1[i][j] == '0' and arr2[i][j] == '0':
                temp += ' '
        # 공백이 이어지는 경우, 하나로 줄이기
        answer.append(' '.join(temp.split()))

    return answer
```

```python
def solution(n, arr1, arr2):
    answer = []
    for i,j in zip(arr1,arr2):
        a12 = str(bin(i|j)[2:])
        a12=a12.rjust(n,'0')
        a12=a12.replace('1','#')
        a12=a12.replace('0',' ')
        answer.append(a12)
    return answer
```
