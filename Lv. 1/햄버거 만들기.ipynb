```python
# 시간 초과

def solution(ingredient):
    answer = 0
    ingredient_change = ''.join(map(str, ingredient))
    while '1231' in ingredient_change:
        index = ingredient_change.find('1231')
        ingredient_change = ingredient_change[:index] + ingredient_change[index+4:]
        answer += 1
    return answer
```

```python
# 실패 [1, 1, 2, 3, 1, 2, 3, 1, 2, 3, 1, 2, 3, 1]
def solution(ingredient):
    answer = 0
    ingredient_change = ''.join(map(str, ingredient))
    while '1231' in ingredient_change:
        answer += ingredient_change.count('1231')
        ingredient_change = ingredient_change.replace('1231', '')
    return answer
```

```python
def solution(ingredient):
    stack = []
    answer = 0

    for num in ingredient:
        stack.append(num)
        if stack[-4:] == [1, 2, 3, 1]:
            answer += 1
            del stack[-4:]

    return answer
```
