```python
def solution(s, n):
    answer = ''
    for alphabet in s:
        if alphabet == ' ':
            answer += ' '
        elif ord(alphabet) >= 65 and ord(alphabet) <= 90:
            if ord(alphabet) + n > 90:
                answer += chr(ord(alphabet) + n - 26)
            else:
                answer += chr(ord(alphabet) + n)
        elif ord(alphabet) >= 97 and ord(alphabet) <= 122:
            if ord(alphabet) + n > 122:
                answer += chr(ord(alphabet) + n - 26)
            else:
                answer += chr(ord(alphabet) + n)
    return answer
```

```python
def solution(s, n):
    answer = ''
    for alphabet in s:
        if alphabet == ' ':
            answer += ' '
        else:
            base = 65 if alphabet.isupper() else 97
            if ord(alphabet) + n > base + 25:
                answer += chr(base + (ord(alphabet) + n - base) % 26)
            else:
                answer += chr(ord(alphabet) + n)
    return answer
```
