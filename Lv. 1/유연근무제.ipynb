```python
def solution(schedules, timelogs, startday):
    # 출근 희망 시각 + 10분
    for i in range(len(schedules)):
        schedules[i] += 10
        if schedules[i] % 100 >= 60:
            schedules[i] += 40

    # 직원들이 일주일 동안 출근한 시각을 요일에 맞춰 재정리
    if startday == 1:
        for i in range(len(timelogs)):
            timelogs[i] = timelogs[i][:5]
    elif startday == 2:
        for i in range(len(timelogs)):
            timelogs[i] = timelogs[i][6:] + timelogs[i][:4]
    elif startday == 3:
        for i in range(len(timelogs)):
            timelogs[i] = timelogs[i][5:] + timelogs[i][:3]
    elif startday == 4:
        for i in range(len(timelogs)):
            timelogs[i] = timelogs[i][4:] + timelogs[i][:2]
    elif startday == 5:
        for i in range(len(timelogs)):
            timelogs[i] = timelogs[i][3:] + timelogs[i][:1]
    elif startday == 6:
        for i in range(len(timelogs)):
            timelogs[i] = timelogs[i][2:]
    else:
        for i in range(len(timelogs)):
            timelogs[i] = timelogs[i][1:6]

    # 상품을 받을 직원의 수
    answer = 0
    for i in range(len(schedules)):
        if max(timelogs[i]) <= schedules[i]:
            answer += 1
    return answer
```
