```python
# 시간 초과

def solution(participant, completion):
    for comp in completion:
        if comp in participant:
            participant.remove(comp)
    return participant[0]
```

```python
def solution(participant, completion):
    participant.sort()
    completion.sort()
    participant = list(enumerate(participant))
    completion = list(enumerate(completion))
    for i, j in completion:
        if j != participant[i][1]:
            return participant[i][1]
    return participant[-1][1]
```
