```python
def solution(babbling):
    answer = 0
    possible = ["aya", "ye", "woo", "ma"]
    possible = list(enumerate(possible, start = 1))
    for i in range(len(babbling)):
        for j,k in possible:
            if k in babbling[i]:
                babbling[i] = babbling[i].replace(k, str(j))
    for b in babbling:
        if b.isdigit():
            if all(b[i] != b[i+1] for i in range(len(b) - 1)):  # 연속된 숫자가 없는지 확인
                answer += 1
    return answer
```

```python
def solution(babbling):
    count = 0

    for b in babbling:
        if "ayaaya" in b or "yeye" in b or "woowoo" in b or "mama" in b:
            continue
        if not b.replace("aya", " ").replace("ye", " ").replace("woo", " ").replace("ma", " ").replace(" ", ""):
            count += 1

    return count
```
