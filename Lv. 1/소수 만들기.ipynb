```python
# 소수 확인
def is_prime(n):
    for i in range(2, int(n**(1/2))+1):
    	if n % i == 0:
        	return False
    return True


def solution(nums):
    from itertools import combinations
    answer = 0
    combi = list(combinations(nums, 3))
    for c in combi:
        if is_prime(sum(c)) == True:
            answer += 1
    return answer
```
