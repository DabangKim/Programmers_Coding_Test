```python
def solution(bandage, health, attacks):
    remaining = health
    for i in range(len(attacks)-1):
        # 공격 받은 직후의 체력
        remaining -= attacks[i][1]

        # 체력이 0 이하라면
        if remaining <= 0:
            return -1

        # 회복 시작
        if attacks[i+1][0] - attacks[i][0] > 1:
            time = attacks[i+1][0] - attacks[i][0] - 1
            if time < bandage[0]:
                remaining += bandage[1] * time
            else:
                remaining += bandage[1] * time + bandage[2] * (time // bandage[0])
        # 최대 체력 유지
        if remaining >= health:
            remaining = health

    # 마지막 공격
    return -1 if remaining - attacks[-1][1] <= 0 else remaining - attacks[-1][1]
```
