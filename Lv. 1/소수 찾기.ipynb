```python
def is_prime(n):
    for i in range(2, int(n**(1/2))+1):
    	if n % i == 0:
        	return False
    return True

def solution(n):
    return sum(1 for i in range(2, n+1) if is_prime(i) == True)
```

```python
def solution(n):
    num=set(range(2,n+1))

    for i in range(2,n+1):
        if i in num:
            num-=set(range(2*i,n+1,i))
    return len(num)
```
