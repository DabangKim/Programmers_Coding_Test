```python
def solution(dartResult):
    answer = 0
    temp = []
    temp_check = 0
    # dartResult 분리시키기
    for i in range(len(dartResult)):
        if dartResult[i].isalpha():
            if "*" in dartResult[temp_check:i+1]:
                temp.append('*')
                temp.append(dartResult[temp_check+1:i+1])
                temp_check = i+1
            elif "#" in dartResult[temp_check:i+1]:
                temp.append('#')
                temp.append(dartResult[temp_check+1:i+1])
                temp_check = i+1
            else:
                temp.append(' ')
                temp.append(dartResult[temp_check:i+1])
                temp_check = i+1
    if dartResult[-1] == "*":
        temp.append('*')
    elif dartResult[-1] == "#":
        temp.append('#')
    else:
        temp.append(' ')
    # dartResult 숫자 변환하기(2D -> 4)
    for j in range(1, 6, 2):
        if 'S' in temp[j]:
            temp[j] = int(temp[j][:len(temp[j])-1])
        elif 'D' in temp[j]:
            temp[j] = int(temp[j][:len(temp[j])-1]) ** 2
        elif 'T' in temp[j]:
            temp[j] = int(temp[j][:len(temp[j])-1]) ** 3
    # *와 #에 따라 계산 적용하기
    cal_check = temp[2] + temp[4] + temp[6]
    if cal_check == '   ':
        answer = temp[1] + temp[3] + temp[5]
    elif cal_check == '  *':
        answer = temp[1] + (temp[3] + temp[5]) * 2
    elif cal_check == '  #':
        answer = temp[1] + temp[3] - temp[5]
    elif cal_check == ' * ':
        answer = (temp[1] + temp[3]) * 2 + temp[5]
    elif cal_check == ' **':
        answer = temp[1] * 2 + temp[3] * 4 + temp[5] * 2
    elif cal_check == ' *#':
        answer = (temp[1] + temp[3]) * 2 - temp[5]
    elif cal_check == ' # ':
        answer = temp[1] - temp[3] + temp[5]
    elif cal_check == ' #*':
        answer = temp[1] - temp[3] * 2 + temp[5] * 2
    elif cal_check == ' ##':
        answer = temp[1] - temp[3] - temp[5]
    elif cal_check == '*  ':
        answer = temp[1] * 2 + temp[3] + temp[5]
    elif cal_check == '* *':
        answer = temp[1] * 2 + (temp[3] + temp[5]) * 2
    elif cal_check == '* #':
        answer = temp[1] * 2 + temp[3] - temp[5]
    elif cal_check == '** ':
        answer = temp[1] * 4 + temp[3] * 2 + temp[5]
    elif cal_check == '***':
        answer = temp[1] * 4 + temp[3] * 4 + temp[5] * 2
    elif cal_check == '**#':
        answer = temp[1] * 4 + temp[3] * 2 - temp[5]
    elif cal_check == '*# ':
        answer = temp[1] * 2 - temp[3] + temp[5]
    elif cal_check == '*#*':
        answer = temp[1] * 2 - temp[3] * 2 + temp[5] * 2
    elif cal_check == '*##':
        answer = temp[1] * 2 - temp[3] - temp[5]
    elif cal_check == '#  ':
        answer = -temp[1] + temp[3] + temp[5]
    elif cal_check == '# *':
        answer = -temp[1] + (temp[3] + temp[5]) * 2
    elif cal_check == '# #':
        answer = -temp[1] + temp[3] - temp[5]
    elif cal_check == '#* ':
        answer = (-temp[1] + temp[3]) * 2 + temp[5]
    elif cal_check == '#**':
        answer = -temp[1] * 2 + temp[3] * 4 + temp[5] * 2
    elif cal_check == '#*#':
        answer = (-temp[1] + temp[3]) * 2 - temp[5]
    elif cal_check == '## ':
        answer = -temp[1] - temp[3] + temp[5]
    elif cal_check == '##*':
        answer = -temp[1] - temp[3] * 2 + temp[5] * 2
    elif cal_check == '###':
        answer = -temp[1] - temp[3] - temp[5]
    return answer
```
