```python
# "-1" 찾기
check = []
for i in range(len(park)):
    for j in range(len(park[0])):
        if park[i][j] == "-1":
            check.append((i, j))
check
```

```python
# 가로별로 묶기
width = []
width_num = []

for i in range(len(check)):
    width_num.append(check[i][0])
width_num = list(set(width_num))

for j in width_num:
    width_temp = []
    for k in range(len(check)):
        if check[k][0] == j:
            width_temp.append(check[k])
    width.append(width_temp)
```
