```python
def solution(mats, park):
    # 1. 문자열로 된 park을 2차원 숫자 배열로 변환 (1 = "-1", 0 = 나머지)
    n = len(park)
    m = len(park[0])
    board = [[1 if park[i][j] == "-1" else 0 for j in range(m)] for i in range(n)]

    # 2. 누적합 계산 (prefix sum)
    prefix = [[0]*(m+1) for _ in range(n+1)]
    for i in range(n):
        for j in range(m):
            prefix[i+1][j+1] = (
                board[i][j]
                + prefix[i][j+1]
                + prefix[i+1][j]
                - prefix[i][j]
            )

    # 3. 가장 큰 돗자리부터 시도해서 놓을 수 있는지 확인
    mats.sort(reverse=True)
    for size in mats:
        for i in range(n - size + 1):
            for j in range(m - size + 1):
                total = (
                    prefix[i+size][j+size]
                    - prefix[i][j+size]
                    - prefix[i+size][j]
                    + prefix[i][j]
                )
                if total == size * size:
                    return size  # 가장 큰 크기부터 시도하므로 처음 찾은 게 최대

    return -1  # 아무 것도 못 놓는 경우
```
