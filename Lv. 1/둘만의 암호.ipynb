```python
def solution(s, skip, index):
    alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

    # skip에 있는 알파벳은 제외 및 z를 넘어갈 경우 a로 돌아가기 위함
    for a in alphabet:
        if a in skip:
            alphabet.remove(a)
    alphabet = alphabet * 2
    alphabet = list(enumerate(alphabet, start=1))

    # 번호 찾기
    temp = []
    for word in s:
        for i,j in alphabet:
            if word == j:
                temp.append(i+index)
                break

    # 정답
    answer = ''
    for i in temp:
        answer += alphabet[i-1][1]
    return answer
```

```python
def solution(s, skip, index):
    # 알파벳 리스트에서 skip 문자를 제외한 새로운 리스트 생성
    alphabet = [chr(i) for i in range(ord('a'), ord('z') + 1) if chr(i) not in skip]

    # 알파벳을 딕셔너리 형태로
    alphabet_dict = {char: idx for idx, char in enumerate(alphabet)}
    alphabet_length = len(alphabet)

    # 정답
    answer = ''
    for word in s:
        # 현재 문자의 인덱스를 찾고, index를 더한 뒤 알파벳 범위 내로 돌려줌
        new_idx = (alphabet_dict[word] + index) % alphabet_length
        answer += alphabet[new_idx]
    return answer
```
