```python
def solution(s):
    for word in s.split():
        temp = ''
        for i in range(len(word)):
            if i % 2 == 0:
                temp += word[i].upper()
            else:
                temp += word[i].lower()
        s = s.replace(word, temp)
    return s
```

```python
solution("ab abcd ab")
```
