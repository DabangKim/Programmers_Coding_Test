```python
def solution(d, budget):
    total = 0
    answer = 0
    i = 0
    d.sort()
    while total < budget:
        total += d[i]
        answer += 1
        i += 1
    return answer - 1
```

```python
def solution(d, budget):
    d.sort()
    total = 0
    for answer, cost in enumerate(d):
        total += cost
        if total > budget:
            return answer
    return len(d)
```
