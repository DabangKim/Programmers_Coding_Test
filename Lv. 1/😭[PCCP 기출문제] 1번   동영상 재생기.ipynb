```python
def solution(video_len, pos, op_start, op_end, commands):

    # 초로 변환
    video_len = int(video_len.split(':')[0]) * 60 + int(video_len.split(':')[1])
    pos = int(pos.split(':')[0]) * 60 + int(pos.split(':')[1])
    op_start = int(op_start.split(':')[0]) * 60 + int(op_start.split(':')[1])
    op_end = int(op_end.split(':')[0]) * 60 + int(op_end.split(':')[1])

    for command in commands:
        # 오프닝 건너뛰기(1)
        if pos >= op_start and pos < op_end:
                pos = op_end

        # 10초 후로 이동
        if command == "next":
            pos += 10
            # 동영상의 남은 시간이 10초 미만일 경우
            if pos > video_len:
                pos = video_len

        # 10초 전로 이동
        else:
            pos -= 10
            # 현재 위치가 10초 미만인 경우
            if pos < 0:
                pos = 0

        # 오프닝 건너뛰기(2)
        if pos >= op_start and pos < op_end:
                pos = op_end

    # 동영상의 위치
    if pos // 60 >= 10:
        if pos % 60 >= 10:
            return str(pos // 60) + ":" + str(pos % 60)
        else:
            return str(pos // 60) + ":0" + str(pos % 60)
    else:
        if pos % 60 >= 10:
            return "0" + str(pos // 60) + ":" + str(pos % 60)
        else:
            return "0" + str(pos // 60) + ":0" + str(pos % 60)
```
