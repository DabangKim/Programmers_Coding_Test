```python
def solution(n, w, num):
    block = []
    line = 0
    answer = 1

    # 상자의 높이 계산
    if n % w == 0:
        line = n // w
    else:
        line = n // w + 1

    # 상자 배열하기
    for i in range(1, line+1):
        temp = [j for j in range((i-1)*w+1, i*w+1)]
        if i % 2 != 0:
            block.append(temp)
        else:
            block.append(list(reversed(temp)))
    block.reverse()

    # n 초과의 수, 0으로 대체
    block = [[value if value <= n else 0 for value in row] for row in block]

    # 꺼내려는 택배 상자의 위치
    for a in range(len(block)):
        if num in block[a]:
            loc = (a, block[a].index(num))

    # 정답
    for i in range(loc[0]):
        if block[i][loc[1]] != 0:
            answer += 1
    return answer
```
