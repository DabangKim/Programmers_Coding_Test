```python
# 시간 초과
# stage.remove(), 리스트를 순차적으로 검색해야 함

def solution(N, stages):
    failure = []
    for i in range(1, N+1):
        clear_X = stages.count(i)
        clear_O = len(stages)
        failure.append(clear_X / clear_O)
        while i in stages:
            stages.remove(i)
    failure = sorted(list(enumerate(failure, start = 1)), key = lambda x:x[1], reverse = True)
    return [i for i, j in failure]
```

```python
def solution(N, stages):
    failure = []
    total_players = len(stages)  # 전체 플레이어 수

    for i in range(1, N+1):
        clear_X = stages.count(i)

        if total_players > 0:
            failure_rate = clear_X / total_players
        else:
            failure_rate = 0

        # 실패율과 단계 정보를 저장
        failure.append((i, failure_rate))

        # 해당 단계에 도달한 플레이어는 제외
        total_players -= clear_X

    # 실패율에 따라 내림차순 정렬
    failure.sort(key=lambda x: x[1], reverse=True)

    # 단계만 반환
    return [stage for stage, rate in failure]
```
