```python
# 시간 초과

def solution(X, Y):
    JJAK = []
    X = list(X)
    Y = list(Y)
    for x in X:
        if x in Y:
            JJAK.append(x)
            Y.remove(x)
    JJAK.sort(reverse=True)
    if JJAK == []:
        return '-1'
    elif JJAK.count('0') == len(JJAK):
        return '0'
    else:
        return ''.join(JJAK)
```

```python
def solution(X, Y):
    X_count = list(set(X))
    Y_count = list(set(Y))
    answer = ''
    for x in X_count:
        if x in Y_count:
            answer += x * min(X.count(x), Y.count(x))
    answer = list(answer)
    answer.sort(reverse=True)
    if answer == []:
        return '-1'
    elif answer.count('0') == len(answer):
        return '0'
    else:
        return ''.join(answer)
```
