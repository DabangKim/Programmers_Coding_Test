```python
def solution(keymap, targets):
    answer = []
    # 입력하려는 문자열부터 확인
    for target in targets:
        press = 0  # press는 문자열을 입력하기 위하여 최소 몇 번씩 눌러야 하는지
        for i in range(len(target)):
            temp = []  # temp는 문자열의 첫글자를 입력하기 위해 몇 번 눌러야하는지 확인하기 위한 리스트
            for key in keymap:
                if target[i] in key:
                    temp.append(key.index(target[i])+1)
                else:
                    temp.append(0)
            # 0은 문자열을 입력할 수 없다는 의미
            # 따라서 temp의 모든 원소가 0이면 [-1]을 return
            # 0이 하나도 없다면 최소한의 누르는 횟수 press에 더하기
            # 0이 하나라도 있으면 전부 제거 후 최소한의 누르는 횟수 press에 더하기
            if temp.count(0) == len(temp):
                press = -1
                break
            elif 0 not in temp:
                press += min(temp)
            else:
                while 0 in temp:
                    temp.remove(0)
                press += min(temp)
        answer.append(press)
    return answer
```
