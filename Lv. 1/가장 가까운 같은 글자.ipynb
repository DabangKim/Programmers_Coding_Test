```python
def solution(s):
    listed_s = list(s)  # 문자를 찾은 후, 0으로 바꾸기 위함
    check = ''          # 문자 있는지 없는지 확인용
    answer = []
    for i in range(len(s)):
        if s[i] not in check:  # 문자가 check 안에 있는지 확인, 없으면 첫 등장한 것이므로 answer에 -1 추가
            check += s[i]
            answer.append(-1)
        else:
            check_point = listed_s.index(s[i])  # check에 있는 문자라면, 처음 등장한 위치를 반환
            answer.append(i - check_point)      # 두 문자 사이의 거리를 answer에 추가
            listed_s[check_point] = 0           # 처음 등장한 문자를 0으로 치환
    return answer
```

```python
def solution(s):
    answer = []
    dic = dict()
    for i in range(len(s)):
        if s[i] not in dic:
            answer.append(-1)
        else:
            answer.append(i - dic[s[i]])
        dic[s[i]] = i

    return answer
```
