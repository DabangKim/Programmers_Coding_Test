```python
def solution(strings, n):
    # n번째 글자 추출 후 번호 붙이기
    numbered_strings = list(enumerate([string[n] for string in strings]))

    # n번째 글자 중복 없이 순서대로 나열
    checklist = list(set([j for i, j in numbered_strings]))
    checklist.sort()

    # 문자가 같을 경우를 위한 순서 배열
    location = []
    for alphabet in checklist:
        temp = []
        for i, j in numbered_strings:
            if alphabet == j:
                temp.append(i)
        location.append(temp)

    # 정답
    answer = []
    for loc in location:
        loc_order = [strings[i] for i in loc]
        for i in sorted(loc_order):
            answer.append(i)

    return answer
```
