```python
# 시간 복잡도 실패 1

def divisor_count(num):
    answer = []
    for i in range(1, num+1):
        if num % i == 0:
            answer.append(i)
    return len(answer)

def solution(number, limit, power):
    attack = []
    for i in range(1, number+1):
        attack.append(divisor_count(i))
    for j in range(len(attack)):
        if attack[j] > limit:
            attack[j] = power
    return sum(attack)
```

```python
# 시간 복잡도 실패 2

def power_limit(num, lim, pow):
    temp = []
    for i in range(1, num+1):
        if num % i == 0:
            temp.append(i)
    if len(temp) > lim:
        return pow
    else:
        return len(temp)

def solution(number, limit, power):
    return sum(power_limit(i, limit, power) for i in range(1, number+1))
```

```python
# 성공

# 약수의 개수를 미리 배열에 저장
def divisor_count(num):
    divisors = [0] * (num + 1)  # 모든 숫자의 약수 개수를 저장할 배열

    for i in range(1, num+1):
        for j in range(i, num+1, i):  # i의 배수들에 대해
            divisors[j] += 1          # j는 i로 나눠떨어지므로, 약수 개수를 +1 증가

    return divisors

def solution(number, limit, power):
    temp = divisor_count(number)
    return sum(power if temp[i] > limit else temp[i] for i in range(1, number + 1))
```
