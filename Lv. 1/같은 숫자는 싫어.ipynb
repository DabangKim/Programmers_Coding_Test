```python
def solution(arr):
    temp = [arr[0]]
    for i in range(len(arr)-1):
        if arr[i] != arr[i+1]:
            temp.append(arr[i+1])
    return temp
```

```python
def solution(arr):
    answer = [arr[0]]
    for a in arr:
        if not a == answer[-1]:
            answer.append(a)
    return answer
```
