```python
# 시간 초과
# 1. 과도한 set 연산
# 2. sorted()

# 칠해지는 구간 미리 배열에 저장
def coloured(start, length):
    return [i for i in range(start, start+length)]

def solution(n, m, section):
    answer = 0
    while section:
        s = section[0]
        intersection = list(set(section) & set(coloured(s, m)))
        section = sorted(list(set(section) - set(intersection)))
        answer += 1
    return answer
```

```python
def solution(n, m, section):
    answer = 0
    i = 0
    while i < len(section):
        answer += 1

        # 현재 칠하는 범위의 끝을 구함
        end = section[i] + m - 1

        # end를 넘지 않는 구간까지 인덱스를 증가시킴
        while i < len(section) and section[i] <= end:
            i += 1

    return answer
```
