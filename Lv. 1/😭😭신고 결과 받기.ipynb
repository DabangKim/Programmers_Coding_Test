```python
def solution(id_list, report, k):
    # 중복 제거
    report = list(set(report))

    # 신고 당한 사람과 횟수
    report_num = [rep.split()[1] for rep in report]
    reported = {name: report_num.count(name) for name in set(report_num)}

    # k번 이상 신고당하면 이용 정지
    stop_id = [name for name, count in reported.items() if count >= k]

    # result 반환
    result = [0] * len(id_list)
    for rep in report:
        if rep.split()[1] in stop_id:
            result[id_list.index(rep.split()[0])] += 1

    return result
```
