```python
def solution(survey, choices):
    scores = {t: 0 for t in "RTCFJMAN"}
    types = [('R', 'T'), ('C', 'F'), ('J', 'M'), ('A', 'N')]

    for s, c in zip(survey, choices):
        score = c - 4
        if score:
            scores[s[score > 0]] += abs(score)

    return ''.join(b if scores[b] > scores[a] else a for a, b in types)
```
