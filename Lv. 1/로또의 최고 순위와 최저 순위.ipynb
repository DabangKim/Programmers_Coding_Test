```python
def solution(lottos, win_nums):
    how_many = 0
    for lotto in lottos:
        if lotto in win_nums:
            how_many += 1
    # 최고 순위
    if lottos.count(0) >= 1:
        highest_ranking = 7 - (how_many + lottos.count(0))
    else:
        if how_many == 0:
            highest_ranking = 6
        else:
            highest_ranking = 7 - how_many
    # 최저 순위
    if how_many <= 0:
        lowest_ranking = 6 - how_many
    else:
        lowest_ranking = 7 - how_many
    return [highest_ranking, lowest_ranking]
```
