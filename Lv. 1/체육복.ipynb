```python
# 실패
def solution(n, lost, reserve):
    lost.sort()
    reserve.sort()

    for l in lost:
        if l in reserve:
            lost.remove(l)
            reserve.remove(l)

    for l in lost:
        if l-1 not in reserve and l+1 not in reserve:
            n -= 1
        elif l-1 in reserve:
            reserve.remove(l-1)
        elif l+1 in reserve:
            reserve.remove(l+1)
    return n
```

```python
def solution(n, lost, reserve):
    # 중복 제거 (여벌이 있지만 도난당한 경우)
    lost_set = set(lost) - set(reserve)
    reserve_set = set(reserve) - set(lost)

    # 체육복을 빌려줄 수 있는 경우 처리
    for l in sorted(lost_set):    # 정렬된 리스트 순회
        if l-1 in reserve_set:    # 앞 번호 학생이 체육복을 빌려줄 수 있는 경우
            reserve_set.remove(l-1)
        elif l+1 in reserve_set:  # 뒷 번호 학생이 체육복을 빌려줄 수 있는 경우
            reserve_set.remove(l+1)
        else:
            n -= 1  # 체육복을 못 빌리는 경우

    return n
```
