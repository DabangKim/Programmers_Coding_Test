```python
def solution(s):
    s = list(s)
    answer = 1

    while len(s) > 2:
        cnt_1 = 0
        cnt_2 = 0
        for i in range(len(s)):
            if s[i] == s[0]:
                cnt_1 += 1
            else:
                cnt_2 += 1
            if cnt_1 == cnt_2:
                answer += 1
                break
        for j in range(cnt_1 + cnt_2):
            s.pop(0)

    return answer
```

```python
solution("abab")
```

```python
def solution(s):
    answer = 0
    i = 0

    while i < len(s):
        cnt_1 = 0
        cnt_2 = 0
        first = s[i]

        while i < len(s):
            if s[i] == first:
                cnt_1 += 1
            else:
                cnt_2 += 1

            i += 1

            if cnt_1 == cnt_2:
                break

        answer += 1

    return answer
```
