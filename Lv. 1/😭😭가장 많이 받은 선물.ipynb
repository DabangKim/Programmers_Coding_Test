```python
# 선물 지수
def index(friends, gifts):
    index = []
    gifts_splited = [gift.split() for gift in gifts]
    for friend in friends:
        gave = 0
        received = 0
        for gift in gifts_splited:
            if friend == gift[0]:
                gave += 1
            if friend == gift[1]:
                received += 1
        index.append((friend, gave-received))
    return index
```
