```python
def solution(friends, gifts):
    n = len(friends)
    name_to_index = {name: i for i, name in enumerate(friends)}

    # A가 B에게 선물을 준 횟수 저장
    gift_count = [[0] * n for _ in range(n)]

    # 총 준 선물과 받은 선물 개수
    total_give = [0] * n
    total_receive = [0] * n

    # 1. 선물 기록 분석
    for gift in gifts:
        giver, receiver = gift.split()
        g_idx = name_to_index[giver]
        r_idx = name_to_index[receiver]
        gift_count[g_idx][r_idx] += 1
        total_give[g_idx] += 1
        total_receive[r_idx] += 1

    # 선물 지수 계산
    gift_score = [total_give[i] - total_receive[i] for i in range(n)]

    # 2. 다음 달 받을 선물 계산
    result = [0] * n  # 다음 달 받을 선물 수

    for i in range(n):
        for j in range(i, n):
            if i == j:
                continue
            # i가 j에게 더 많이 줬는가?
            if gift_count[i][j] > gift_count[j][i]:
                result[i] += 1
            elif gift_count[i][j] < gift_count[j][i]:
                result[j] += 1
            else:
                # 선물 지수 비교
                if gift_score[i] > gift_score[j]:
                    result[i] += 1
                elif gift_score[i] < gift_score[j]:
                    result[j] += 1
                # 같으면 아무도 안 줌

    return max(result)
```
