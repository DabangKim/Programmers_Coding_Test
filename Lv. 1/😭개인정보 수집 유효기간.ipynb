```python
def solution(today, terms, privacies):

    # 변수 치환

    splited_today = ""
    splited_terms = []
    splited_privacies_v1 = []
    splited_privacies_v2 = []

    splited_today = today.split('.')

    for term in terms:
        splited_terms.append(term.split(' '))

    for privacy in privacies:
        splited_privacies_v1.append(privacy.split(' '))
    for privacy in splited_privacies_v1:
        splited_privacies_v2.append([privacy[0].split('.'), privacy[1]])

    # 정답
    answer = []
    i = 1
    for privacy in splited_privacies_v2:
        for term in splited_terms:

            # 맞는 약관 찾기
            if term[0] == privacy[1]:

                # 날짜 계산하기
                year = int(splited_today[0]) - int(privacy[0][0])
                month = int(splited_today[1]) - int(privacy[0][1])
                day = int(splited_today[2]) - int(privacy[0][2])
                for_check = year * 12 + month
                if for_check == int(term[1]) and day >= 0:
                    answer.append(i)
                elif for_check > int(term[1]):
                    answer.append(i)
                i += 1

    return answer
```

```python
def to_days(date):
    year, month, day = map(int, date.split("."))
    return year * 28 * 12 + month * 28 + day

def solution(today, terms, privacies):
    months = {v[0]: int(v[2:]) * 28 for v in terms}
    today = to_days(today)
    expire = [
        i + 1 for i, privacy in enumerate(privacies)
        if to_days(privacy[:-2]) + months[privacy[-1]] <= today
    ]
    return expire
```
