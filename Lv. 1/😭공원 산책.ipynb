```python
def solution(park, routes):

    # park를 2차원 배열로 바꾸기
    park_change = []
    for p in park:
        temp = []
        for i in p:
            temp.append(i)
        park_change.append(temp)

    # routes를 문자열로 바꾸기
    routes_change = []
    for route in routes:
        route = route.split(' ')
        temp = ''
        for i in range(int(route[1])):
            temp += route[0]
        routes_change.append(temp)

    # 시작 지점 찾기
    start_point = [0, 0]
    for i in range(len(park_change)):
        for j in range(len(park_change[0])):
            if park_change[i][j] == 'S':
                start_point = [i,j]

    # 최종 위치
    current_point = start_point[:]
    previous_point = start_point[:]
    for route in routes_change:
        go_back = False
        for r in route:
            if r == 'E':
                current_point[1] += 1
                if current_point[1] > len(park_change[0]) - 1 or park_change[current_point[0]][current_point[1]] == 'X':
                    go_back = True
            elif r == 'W':
                current_point[1] -= 1
                if current_point[1] < 0 or park_change[current_point[0]][current_point[1]] == 'X':
                    go_back = True
            elif r == 'S':
                current_point[0] += 1
                if current_point[0] > len(park_change) - 1 or park_change[current_point[0]][current_point[1]] == 'X':
                    go_back = True
            else:
                current_point[0] -= 1
                if current_point[0] < 0 or park_change[current_point[0]][current_point[1]] == 'X':
                    go_back = True
        if go_back == True:
            current_point = previous_point[:]
        else:
            previous_point = current_point[:]
        print(current_point)
    return current_point
```
