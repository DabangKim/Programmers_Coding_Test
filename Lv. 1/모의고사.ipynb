```python
def solution(answers):
    soo_1 = [1, 2, 3, 4, 5]
    soo_2 = [2, 1, 2, 3, 2, 4, 2, 5]
    soo_3 = [3, 3, 1, 1, 2, 2, 4, 4, 5, 5]

    quotient_1 = len(answers) // len(soo_1)
    remainder_1 = len(answers) % len(soo_1)
    soo_1 = soo_1 * quotient_1 + soo_1[:remainder_1]
    quotient_2 = len(answers) // len(soo_2)
    remainder_2 = len(answers) % len(soo_2)
    soo_2 = soo_2 * quotient_2 + soo_2[:remainder_2]
    quotient_3 = len(answers) // len(soo_3)
    remainder_3 = len(answers) % len(soo_3)
    soo_3 = soo_3 * quotient_3 + soo_3[:remainder_3]

    answer_check_1, answer_check_2, answer_check_3 = 0, 0, 0
    for i in range(len(answers)):
        if answers[i] == soo_1[i]:
            answer_check_1 += 1
        if answers[i] == soo_2[i]:
            answer_check_2 += 1
        if answers[i] == soo_3[i]:
            answer_check_3 += 1

    answer = list(enumerate([answer_check_1, answer_check_2, answer_check_3], start = 1))
    answer.sort(key = lambda x : x[1], reverse = True)
    if answer[0][1] != answer[1][1]:
        return [answer[0][0]]
    elif answer[0][1] == answer[1][1] and answer[1][1] != answer[2][1]:
        return [answer[0][0], answer[1][0]]
    else:
        return [answer[0][0], answer[1][0], answer[2][0]]
```
