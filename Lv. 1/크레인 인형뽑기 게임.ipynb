```python
def solution(board, moves):
    # 행렬 전치
    board = [list(x) for x in zip(*board)]
    # 0 제거
    for b in board:
        while 0 in b:
            b.remove(0)
    # 바구니
    baguuni = []
    for m in moves:
        if board[m-1] != []:
            baguuni.append(board[m-1][0])
            board[m-1].pop(0)
    # 정답
    stack = []
    answer = 0
    for b in baguuni:
        stack.append(b)
        if len(stack) >= 2:
            if stack[-2] == stack[-1]:
                answer += 2
                del stack[-2:]
    return answer
```

```python
solution([[0,0,0,0,0],[0,0,1,0,3],[0,2,5,0,1],[4,2,4,4,2],[3,5,1,3,1]], [1,5,3,5,1,2,1,4])
```
