```python
def solution(numLog):
    alLog = []
    answer = ''

    for i in range(len(numLog) - 1):
        alLog.append(numLog[i+1] - numLog[i])

    for i in alLog:
        if i == 1:
            answer += 'w'
        elif i == -1:
            answer += 's'
        elif i == 10:
            answer += 'd'
        else:
            answer += 'a'

    return answer
```
