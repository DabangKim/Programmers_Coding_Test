```python
def solution(arr, queries):
    list = []
    answer = []

    for query in queries:
        for i in range(query[0], query[1]+1):
            if arr[i] > query[2]:
                list.append(arr[i])
        if list == []:
            list.append(-1)
        answer.append(min(list))
        list.clear()

    return answer
```

```python
def solution(arr, queries):
    answer = []
    for s, e, k in queries:
        tmp = []
        for x in arr[s:e+1]:
            if x > k:
                tmp.append(x)
        answer.append(-1 if not tmp else min(tmp))
    return answer
```
