```python
def solution(myString):
    save = [-1]
    result = []
    for i in range(len(myString)):
        if 'x' in myString[i]:
            save.append(i)
    save.append(len(myString))
    for num in range(len(save)-1):
        result.append(save[num+1]-save[num]-1)
    return result
```
