```python
def solution(rank, attendance):
    rank_attend = []
    order = list(enumerate(rank))
    for i in range(len(attendance)):
        if attendance[i] == True:
            rank_attend.append(order[i])
    rank_attend_order = sorted(rank_attend, key = lambda x : x[1])
    return rank_attend_order[0][0] * 10000 + rank_attend_order[1][0] * 100 + rank_attend_order[2][0]
```
