```python
def solution(n):
    # n x n 배열 초기화 (0으로 채움)
    arr = [[0] * n for _ in range(n)]

    # 1부터 n^2까지 채울 숫자
    num = 1
    # 상, 하, 좌, 우의 방향 설정 (시계방향)
    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]  # 오른쪽, 아래, 왼쪽, 위
    x, y = 0, 0  # 시작 위치 (0, 0)
    dir_index = 0  # 현재 방향 (0: 오른쪽, 1: 아래, 2: 왼쪽, 3: 위)

    while num <= n * n:
        # 현재 위치에 num 값 채우기
        arr[x][y] = num
        num += 1

        # 다음 위치 계산
        next_x = x + directions[dir_index][0]
        next_y = y + directions[dir_index][1]

        # 다음 위치가 배열의 범위를 벗어나거나 이미 값이 채워져 있으면 방향 전환
        if not (0 <= next_x < n and 0 <= next_y < n and arr[next_x][next_y] == 0):
            # 방향 전환 (시계방향)
            dir_index = (dir_index + 1) % 4
            next_x = x + directions[dir_index][0]
            next_y = y + directions[dir_index][1]

        # 다음 위치로 이동
        x, y = next_x, next_y

    return arr
```
