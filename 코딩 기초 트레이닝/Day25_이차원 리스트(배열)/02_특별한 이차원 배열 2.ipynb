```python
def solution(arr):
    temp = []
    for i in range(len(arr)):
        for j in range(len(arr)):
            if arr[i][j] == arr[j][i]:
                temp.append(True)
            else:
                temp.append(False)
    if False in temp:
        return 0
    else:
        return 1
```
