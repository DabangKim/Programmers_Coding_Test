```python
def solution(arr):
    if len(arr) > len(arr[0]):
        temp = len(arr) - len(arr[0])
        for i in range(len(arr)):
            for j in range(temp):
                arr[i].append(0)
    elif len(arr) < len(arr[0]):
        temp = len(arr[0]) - len(arr)
        for i in range(temp):
            temp_list = []
            for j in range(len(arr[0])):
                temp_list.append(0)
            arr.append(temp_list)
    return arr
```

```python
def solution(arr):
    n=len(arr)
    m=len(arr[0])
    if n>m:
        for i in range(n):
            for j in range(n-m):
                arr[i].append(0)
    else:
        for i in range(m-n):
            arr.append([0]*m)

    return arr
```
