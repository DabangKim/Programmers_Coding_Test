```python
def solution(order):
    answer = 0
    for menu in order:
        if menu == "iceamericano" or menu == "americanoice" or menu == "hotamericano" or menu == "americanohot" or menu == "americano" or menu == "anything":
            answer += 4500
        elif menu == "icecafelatte" or menu == "cafelatteice" or menu == "hotcafelatte" or menu == "cafelattehot" or menu == "cafelatte":
            answer += 5000
    return answer
```
