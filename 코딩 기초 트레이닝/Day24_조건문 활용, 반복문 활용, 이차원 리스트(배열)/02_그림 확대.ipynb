```python
def solution(picture, k):
    answer = []
    for pic in picture:
        temporary = ''
        for i in pic:
            temporary += i * k
        for num in range(k):
            answer.append(temporary)
    return answer
```

```python
def solution(picture, k):
    answer = []
    for i in range(len(picture)):
        for _ in range(k):
            answer.append(picture[i].replace('.', '.' * k).replace('x', 'x' * k))
    return answer
```
