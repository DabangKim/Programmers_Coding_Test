```python
def solution(arr):
    answer = []
    if 2 not in arr:
        return [-1]
    else:
        list = []
        for i in range(len(arr)):
            if arr[i] == 2:
                list.append(i)
        for num in range(list[0], list[-1]+1):
            answer.append(arr[num])
        return answer
```
