```python
def solution(num_list):
    answer = -1
    for i in range(len(num_list)):
        if num_list[i] < 0:
            answer = i
            break
    return answer
```

```python
def solution(num_list):
    return ([i for i in range(len(num_list)) if num_list[i] < 0] or [-1])[0]
```

```python
def solution(num_list):
    for i, num in enumerate(num_list):
        if num < 0:
            return i
    return -1
```
