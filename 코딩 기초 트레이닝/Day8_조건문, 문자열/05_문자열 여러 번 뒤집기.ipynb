```python
# 실패

def solution(my_string, queries):
    for query in queries:
        list = []
        alphabet = []
        word = ''
        for num in range(query[0], query[1]+1):
            list.append(num)
        for num in list:
            alphabet.append(my_string[num])
        alphabet.reverse()
        for alpha in alphabet:
            word += alpha
        my_string = my_string.replace(my_string[query[0]:query[1]+1], word)
    return my_string
```

```python
# 성공

def solution(my_string, queries):
    for num in range(len(queries)):
        first = queries[num][0]
        second = queries[num][1] + 1
        change = my_string[first:second]
        change = change[::-1]
        a = list(my_string)
        a[first:second] = list(change)
        my_string = ''.join(a)
    return my_string
```

```python
def solution(my_string, queries):
    for (s, e) in queries:
        my_string = my_string[:s] + my_string[s:e+1][::-1] + my_string[e+1:]
    return my_string
```
