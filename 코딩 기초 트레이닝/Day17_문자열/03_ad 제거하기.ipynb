```python
def solution(strArr):
    TF = []
    li = []
    for i in range(len(strArr)):
        if 'ad' not in strArr[i]:
            TF.append(True)
        else:
            TF.append(False)
    for i in range(len(TF)):
        if TF[i] == True:
            li.append(strArr[i])
    return li
```
