```python
def solution(names):
    name_list = []
    # 몇 바퀴 돌지 결정
    if len(names) % 5 == 0:
        num = len(names) // 5
    else:
        num = len(names) // 5 + 1
    for i in range(num):
        name_list.append(names[5*i])
    return name_list
```
