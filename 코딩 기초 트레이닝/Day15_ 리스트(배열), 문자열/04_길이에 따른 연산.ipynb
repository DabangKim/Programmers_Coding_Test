```python
def solution(num_list):
    if len(num_list) >= 11:
        add = 0
        for num in num_list:
            add += num
        return add
    else:
        mul = 1
        for num in num_list:
            mul *= num
        return mul
```

```python
from math import prod
def solution(num_list):
    return sum(num_list) if len(num_list)>=11 else prod(num_list)
```
