```python
def change(arr):
    arr_change = []
    for i in range(len(arr)):
        if arr[i]>=50 and arr[i]%2==0:
            arr_change.append(int(arr[i]/2))
        elif arr[i]<50 and arr[i]%2==1:
            arr_change.append(arr[i]*2+1)
        else:
            arr_change.append(arr[i])
    return arr_change

def solution(arr):
    x = 0
    before = arr
    while True:
        after = change(before)
        if before == after:
            return x
        else:
            before = after
            x += 1
            continue
```

```python
def solution(arr):
    answer = 0
    old = arr
    while(True):
        new = []
        for i in old:
            if i>=50 and i%2 == 0:
                i = i/2
            elif i<50 and i%2 == 1:
                i = i*2 + 1
            new.append(int(i))
        if old == new:
            break
        else:
            old = new
            answer += 1
    return answer
```
