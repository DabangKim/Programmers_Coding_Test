```python
def solution(num_list):
    MUL = 1
    ADD_SQUARE = 0
    for i in num_list:
        MUL *= i
        ADD_SQUARE += i
    ADD_SQUARE = ADD_SQUARE**2
    if MUL > ADD_SQUARE:
        answer = 0
    else:
        answer = 1
    return answer
```
