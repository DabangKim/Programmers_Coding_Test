```python
def solution(num_list):
    str_even = ''
    str_odd = ''
    for num in num_list:
        if num % 2 == 0:
            str_even += str(num)
        else:
            str_odd += str(num)
    str_even = int(str_even)
    str_odd = int(str_odd)
    answer = str_even + str_odd
    return answer
```
