```python
def solution(a, d, included):
    list = []
    sum = 0
    for i in range(len(included)):  # 등차수열 list 생성
        list.append(a + d * i)

    for num in range(len(included)):  # sum 구하기
        if included[num] == True:
            sum += list[num]

    return sum
```
